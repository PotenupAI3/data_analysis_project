```python
import pandas as pd

# 두 데이터 로드
df_mig = pd.read_csv("migration_data.csv")
df_house = pd.read_csv("officetel_data.csv")
```

```python
# 행정코드 타입 통일 (문자열로)
df_mig["행정코드"] = df_mig["행정코드"].astype(str)
df_house["행정코드"] = df_house["행정코드"].astype(str)

# 데이터 병합 (Inner Join)
# '행정코드'를 기준으로 두 데이터를 하나로 합침
df_final = pd.merge(df_house, df_mig, on="행정코드", how="inner")

display(df_final.head())
```
