```python
# 1. 필요한 열만 리스트로 뽑기
needed_cols = ['C1_NM', 'C2_NM', 'ITM_NM', 'PRD_DE', 'DT']

# 2. 해당 열만 남기고 새 변수에 담기 (나머지는 자동으로 숨겨짐/삭제됨)
df_ana = df[needed_cols].copy()
```

```python
# 3. ITM_NM이 '인구(내국인)'인 것만 남기기
df_ana = df_ana[df_ana['ITM_NM'] == '인구(내국인)']

# 4. C2_NM에서 청년층(예: 20~34세)에 해당하는 값만 필터링
youth_labels = ['20~24세', '25~29세', '30~34세']
df_ana = df_ana[df_ana['C2_NM'].isin(youth_labels)]
```
