```python
import os

from dotenv import load_dotenv

load_dotenv()

key = os.getenv("KOSIS_API_KEY")

print(key)
```

```python
import requests as req

base_url = "https://kosis.kr/openapi/statisticsData.do"

"""
https://kosis.kr/openapi/statisticsData.do
?method=getMeta&type=ITM
&apiKey=YOUR_KOSIS_KEY
&orgId=101
&tblId=DT_1YL20642
&format=json
"""

url = """https://kosis.kr/openapi/statisticsData.do?method=getMeta&type=ITM
&apiKey=Y{key}
&orgId=101
&tblId=DT_1YL20642
&format=json
"""

pararms = {
    "method": "getMeta",
    "type": "ITM",
    "apiKey": key,
    "orgId": "101",
    "tblId": "DT_1YL20642",
    "format": "json",
}


# res = req.get(url)
res = req.get(base_url, params=pararms)

print(res.status_code)
```
