```python
import numpy as np
from PIL import Image

img = Image.open("../../images/mask_map.png").convert("L")
arr = np.array(img)

thr = 230  # 200~245 사이에서 조절
mask = np.where(arr < thr, 0, 255).astype(np.uint8)

print(mask.min(), mask.max())
print(np.unique(mask))
```
