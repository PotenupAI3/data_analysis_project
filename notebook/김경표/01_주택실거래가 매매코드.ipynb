```python
import pandas as pd

df = pd.read_csv("../../data/국토교통부_주택실거래가 매매코드_20201119.csv", encoding="cp949")
df.head()
```

```python
df["구축년도"] = pd.to_datetime(df["구축년도"])
df.head()
```

```python
df["기타주소 건물명"] = df["기타주소 건물명"].fillna("(0-0)")
```

```python
df = df.loc[:, ["법정동코드", "행정동코드", "법정동주소"]]
df["지역코드"] = df["법정동코드"] // 100000
df
```

```python

df.groupby(by="지역코드")["법정동주소"].value_counts()

```

```python
df['시군구명'] = df['법정동주소'].apply(lambda x: ' '.join(x.split()[:2]))
df
```

```python
df = df.loc[:, ["시군구명", "지역코드"]]
df.groupby("시군구명")["지역코드"].unique()
df
```

```python
# 1. 지역코드와 시군구명 컬럼만 선택해서 중복 제거
unique_area = df[['지역코드', '시군구명']].drop_duplicates()

# 2. 지역코드 순으로 정렬 (보기 편하게)
unique_area = unique_area.sort_values('지역코드')
df = unique_area.reset_index(drop=True)

# 3. 결과 확인
df

```

```python
df.to_csv("../../data/국토교통부_주택실거래가 매매코드_20201119_clean.csv")
```
